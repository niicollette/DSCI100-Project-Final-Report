3.4 Model Comparison & Interpretation

In this section, we compare and interpret the three analysis methods used in this project, including linear regression, K near-neighbor regression (KNN) and statistical inference methods (t-test and single-factor analysis of variance ANOVA) to evaluate their effects in interpreting and predicting the duration of players' games.

Comparison of the performance of prediction models

Both prediction models - linear regression and KNN regression - have limited performance in predicting the duration of a single game session. The decision coefficient (R²) of the linear regression model is close to zero, indicating that the two time variables of login time and day of the week can hardly explain the variability of the duration of the session. Although the linear regression model has good interpretability, its prediction error (MAE and RMSE) is still large, indicating that its prediction ability is weak.

In contrast, the performance of the KNN regression model is worse, its error index is higher, and R² is negative. This shows that the prediction effect of the model on the test data is even worse than the simple mean prediction. One possible reason is that the duration of the game session itself is highly random, and it is difficult to find stable similar neighbors by relying only on time-related characteristics. In addition, the KNN method is sensitive to feature selection and distance measurement, which further limits its application effect in this scenario.

Generally speaking, whether you use a parametric model or a non-parametric model, it is difficult to accurately predict the duration of a single session using only time characteristics such as login time and day of the week.


Statistical inference and analysis results

Unlike the predictive model, the method based on statistical inference reveals more meaningful group-level behavior patterns. By comparing the session duration of login in the morning and afternoon/evening through the independent sample t test, we did not observe a significant difference. However, the results of single-factor analysis of variance (ANOVA) on different dates of the week show that there is a significant difference in the duration of the session between weeks, and the p value is much smaller than the commonly used significance level.

This result highlights the important difference between statistical significance and predictability. Although the day of the week has a significant impact on the duration of the session at the group average level, this impact is not enough to support the accurate prediction of the duration of a single session. In other words, statistical inference methods are suitable for revealing overall behavioral trends, rather than accurate predictions at the individual level.

Explanation of results and practical significance

The contrast between the weak performance of the prediction model and the significant statistical inference results shows that the duration of the game session may be affected by a certain time law and a large number of unobserved factors at the same time. Although there are differences in the average behavior of players on different dates, the duration of a single game may still be affected by factors such as personal free time, social interaction and game goals, which are not included in the current data set.

Therefore, time variable-based analysis is valuable in describing and explaining the overall player behavior trend, but it is not suitable for accurately predicting the session duration of a single player. For the research team in charge of server resource planning, instead of trying to predict the specific session duration, it is better to focus on the overall pattern of player activities or adopt a coarser classification method to more effectively support resource allocation decision-making.
